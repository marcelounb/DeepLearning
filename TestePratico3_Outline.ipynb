```
%tensorflow_version 2.x
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random 
import cv2
from cv2 import cvtColor
from google.colab.patches import cv2_imshow

tf.__version__
```

```
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras import utils
```

```
from PIL import Image, ImageFilter
```

```
img_orig = Image.open("/content/dudu.jpg")
img_orig
```

```
kernel_emboss = [-2,-1,0,-1,1,1,0,1,2]  # emboss
```

```
 img2 = img_orig.filter(ImageFilter.Kernel((3,3), kernel_emboss))
img2
```

```
img3 = img_orig.filter(ImageFilter.BLUR)
img3
```

```
img4 = img_orig.filter(ImageFilter.CONTOUR)
img4
```

```
img5 = img_orig.filter(ImageFilter.FIND_EDGES)
img5
```

```
img6 = img_orig.filter(ImageFilter.SHARPEN)
img6
```

```
img7 = img_orig.filter(ImageFilter.SMOOTH_MORE)
img7
```

```
img8 = img_orig.convert('LA')
img8
```

```
kernel_custom_contour = [-1, -1, -1, -1, 11, -2, -2, -2, -2]  # customized contour
img9 = img_orig.filter(ImageFilter.Kernel((3,3), (kernel_custom_contour),1,0))
img9
```

```
images = []
images.append('empty')
images.append(img_orig)
images.append(img2)
images.append(img3)
images.append(img4)
images.append(img5)
images.append(img6)
images.append(img7)
images.append(img8)
images.append(img9)
```

```
legenda = []
legenda.append('Empty')
legenda.append('Original')
legenda.append('emboss')
legenda.append('BLUR')
legenda.append('CONTOUR')
legenda.append('FIND_EDGES')
legenda.append('SHARPEN')
legenda.append('SMOOTH_MORE')
legenda.append('GrayScale')
legenda.append('Custom Contour')

```

```
plt.figure(figsize=(25,25))
for n in range(len(images)-1):
  plt.subplot(6,6,n+1)
  plt.imshow(images[n+1])
  plt.title(legenda[n+1])
  plt.axis('off')
```
