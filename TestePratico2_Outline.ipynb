```
%tensorflow_version 2.x
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random 
import cv2
from google.colab.patches import cv2_imshow

tf.__version__
```

```
img = cv2.imread('/content/dudu.jpg')
img2 = cv2.imread('/content/mini.jpg')
cv2_imshow(img)
```

```

img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY).astype(float)

edge_x = cv2.Sobel(img,cv2.CV_64F,1,0,ksize = 3)
edge_y = cv2.Sobel(img,cv2.CV_64F,0,1,ksize = 3)
edge = np.sqrt(edge_x ** 2 + edge_y ** 2)

# image can be normalized to
# fit into 0..255 color space
```

```
cv2_imshow(img)
```

```
cv2.imwrite('dudu_convol.jpg', edge)
cv2_imshow(edge)
```
